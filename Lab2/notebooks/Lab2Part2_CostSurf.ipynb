```python
#build the extract part of the ETL to MN DNR

from ftplib import FTP
ftp = FTP ('ftp.lmic.state.mn.us')
ftp.login()
ftp.cwd("/pub/data/elevation/lidar/county/winona/laz/")
ftp.retrlines('LIST')
```

```python
#download the data as a zipfile and unzip for use in Pro

with open('elevation_data.gdb.zip', 'wb') as fp:
    ftp.retrbinary('RETR elevation_data.gdb.zip' , fp.write)
```

```python
from zipfile import ZipFile

# Create a ZipFile Object and load sample.zip in it
with ZipFile('elevation_data.gdb.zip', 'r') as zipObj:
   # Extract all the contents of zip file in current directory
   zipObj.extractall()
```
